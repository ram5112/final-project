```python
data_set.drop(['reviews.userCity'],axis = 1)  #https://www.geeksforgeeks.org/how-to-drop-one-or-multiple-columns-in-pandas-dataframe/
data_set.drop(['reviews.userProvince'],axis = 1)
data_set.drop(['colors'],axis = 1)
data_set.drop(['dimension'],axis = 1)
data_set.drop(['sizes'],axis = 1)
data_set.drop(['upc'],axis = 1)
data_set.drop(['ean'],axis = 1)
data_set.drop(['reviews.userProvince'],axis = 1)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
print("Before {}".format(len(data_set)))
dataAfter = data_set.dropna(subset=["reviews.rating"])
# Removes all NAN in reviews.rating
print("After {}".format(len(dataAfter)))
dataAfter["reviews.rating"] = dataAfter["reviews.rating"].astype(int)
```

```python
def sentiments(rating):
    if (rating == 5) or (rating == 4):
        return "positive"
    elif rating == 3:
        return "neutral"
    elif (rating == 2) or (rating == 1):
        return "negetive"
# Add sentiments to the data
strat_train["Sentiment"] = strat_train["reviews.rating"].apply(sentiments)
strat_test["Sentiment"] = strat_test["reviews.rating"].apply(sentiments)
strat_train["Sentiment"][:20]
```

```python
X_train = strat_train["reviews.text"]
X_train_targetSentiment = strat_train["Sentiment"]
X_test = strat_test["reviews.text"]
X_test_targetSentiment = strat_test["Sentiment"]
print(len(X_train), len(X_test))
```

```python
X_train = X_train.fillna(' ')
X_test = X_test.fillna(' ')
X_train_targetSentiment = X_train_targetSentiment.fillna(' ')
X_test_targetSentiment = X_test_targetSentiment.fillna(' ')
# Text preprocessing and occurance counting
from sklearn.feature_extraction.text import CountVectorizer 
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train) 
X_train_counts.shape
```

```python
X_train_targetSentiment
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer(use_idf=False)
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

```python
training_times = []
```

```python
import time
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
clf_multiNB_pipe = Pipeline([("vect", CountVectorizer()), 
                             ("tfidf", TfidfTransformer()),
                             ("clf_nominalNB", MultinomialNB(alpha=0.5, fit_prior=True, class_prior=None))])
start = time.time()
clf_multiNB_pipe.fit(X_train, X_train_targetSentiment)
stop = time.time()
NBtrain = stop - start
training_times.append(NBtrain)
```

```python
import numpy as np
predictedMultiNB = clf_multiNB_pipe.predict(X_test)
np.mean(predictedMultiNB == X_test_targetSentiment)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
clf_logReg_pipe = Pipeline([("vect", CountVectorizer()), 
                            ("tfidf", TfidfTransformer()), 
                            ("clf_logReg", LogisticRegression())])
start = time.time()
clf_logReg_pipe.fit(X_train, X_train_targetSentiment)
stop = time.time()
LRtrain =stop - start
training_times.append(LRtrain)
import numpy as np
predictedLogReg = clf_logReg_pipe.predict(X_test)
np.mean(predictedLogReg == X_test_targetSentiment)
```

```python
import time
from sklearn.svm import LinearSVC
clf_linearSVC_pipe = Pipeline([("vect", CountVectorizer()), 
                               ("tfidf", TfidfTransformer()),
                               ("clf_linearSVC", LinearSVC(max_iter=10000 , verbose =2, C=1))])
start = time.time()
clf_linearSVC_pipe.fit(X_train, X_train_targetSentiment)
stop = time.time()
SVCtime= stop - start
training_times.append(SVCtime)
predictedLinearSVC = clf_linearSVC_pipe.predict(X_test)
np.mean(predictedLinearSVC == X_test_targetSentiment)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_decisionTree_pipe = Pipeline([("vect", CountVectorizer()), 
                                  ("tfidf", TfidfTransformer()), 
                                  ("clf_decisionTree", DecisionTreeClassifier(max_depth = 100))
                                 ])
start = time.time()
clf_decisionTree_pipe.fit(X_train, X_train_targetSentiment)
stop = time.time()
DTtime = stop - start
training_times.append(DTtime)
predictedDecisionTree = clf_decisionTree_pipe.predict(X_test)
np.mean(predictedDecisionTree == X_test_targetSentiment)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_randomForest_pipe = Pipeline([("vect", CountVectorizer()), 
                                  ("tfidf", TfidfTransformer()), 
                                  ("clf_randomForest", RandomForestClassifier(n_estimators = 1000))
                                 ])
start = time.time()
clf_randomForest_pipe.fit(X_train, X_train_targetSentiment)
stop = time.time()
RFtime = stop - start
training_times.append(RFtime)
predictedRandomForest = clf_randomForest_pipe.predict(X_test)
np.mean(predictedRandomForest == X_test_targetSentiment)
```

```python
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn5 = KNeighborsClassifier(n_neighbors = 3)
from sklearn.pipeline import Pipeline
clf_Knn_pipe = Pipeline([("vect", CountVectorizer()), 
                             ("tfidf", TfidfTransformer()),
                             ("clf_Knn",KNeighborsClassifier(n_neighbors = 3))])
start = time.time()
clf_Knn_pipe.fit(X_train, X_train_targetSentiment)
stop = time.time()
Knntime = stop - start
training_times.append(Knntime)
```

```python
import numpy as np
predictedKnn = clf_Knn_pipe.predict(X_test)
np.mean(predictedKnn == X_test_targetSentiment)
```
